### Edge Detection

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('Data/dog_face.jpg')
```

```python
plt.imshow(img)
```

```python
edges = cv2.Canny(image=img, threshold1=127, threshold2=127)
```

```python
plt.imshow(edges)
```

```python
edges = cv2.Canny(image=img, threshold1=0, threshold2=255)
```

```python
plt.imshow(edges)
```

### Threshold Seimi
---
https://stackoverflow.com/questions/25125670/best-value-for-threshold-in-canny

```python
# Pixel deerlerinin mediann hesaplyoruz
med_val = np.median(img) 
```

```python
med_val
```

```python
# Lower deer iin optimum deeri bulmaya alyoruz
lower = int(max(0, 0.7* med_val))
```

```python
# Upper deer iin optmum deeri bulmay alyoruz
upper = int(min(255,1.3 * med_val))
```

```python
edges = cv2.Canny(image=img, threshold1=lower , threshold2=upper)
```

```python
plt.imshow(edges)
```

*Blurred image bize bu noktada yardm edebilir.*

```python
blurred_img = cv2.blur(img,ksize=(5,5))
```

```python
edges = cv2.Canny(image=blurred_img, threshold1=lower , threshold2=upper)
```

```python
plt.imshow(edges)
```
